# Fetching all the tables

```python

def fetch_tables(connection): # this fetch tables function will take a connection object as parameter
    cursor=connection.cursor() # after connectionn has been established we will create a cursro which will point to the connection
    try:
        cursor.execute("SHOW TABLES") # show table is a query which this cursor/ pointer is running
        tables=cursor.fetchall() # to fetch all the tables
        return tables
    except Exception as e:
        print(f"The error '{e}' has been occured.")
        return None

# fetching all the tables
if connection: # if we get connected
    tables=fetch_tables(connection) # this function will be called
    if tables:
        print("Tables in the database is: ")
        for table in tables:
            print(table)
```

# creating a new table

```python
# A COMMON QUERY FOR ALL THE BASIC COMMAND
def executeQuery(connection, query):
    cursor=connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query commited succesfully")
    except Exception as e:
        print(f"The error '{e}' has been occured")

# creating a new table
createNewTable="CREATE TABLE IF NOT EXISTS test_tableI(id INT AUTO_INCREMENT,name TEXT NOT NULL,age INT, PRIMARY KEY (id)) ENGINE = InnoDB"
# above command is the sql command wrapped up in a variable and
# this variable will be passed in further query execution function

executeQuery(connection, createNewTable)
```

# Inserting data into the table

```python
insertData= "INSERT INTO test_tableI (name, age ) VALUES ('Alice', 21 ), ('John', 35)"
executeQuery(connection, insertData) 
```

# Reading data from a table

```python

def readTable(connection, query):
    cursor=connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result= cursor.fetchall()
        return result
    except Exception as e:
        print(f"The error is: {e}")


# now reading that data from the table
select_data="SELECT * FROM test_tableI"
data= readTable(connection, select_data)

print("Data in the test table: ")
for row in data:
    print(row)
```
