# upsampling, downsampling


```python
# we need to generate a random data set -> import numpy and pandas
import numpy as np
import pandas as pd




# we will create two class dataset
nSample= 1000
class0ratio=0.9
nClass0= int(nSample * class0ratio)
nClass1=nSample - nClass0

nClass0, nClass1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
class0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=nClass0),
    'feature2': np.random.normal(loc=0, scale=1, size=nClass0),
    'target': [0] * nClass0 # number of zero = 900
})
# we will create two dataframes and do normal distribution
# feature-> random normal distributon; location~ mean, scale~ standard deviation, size: 900 for class0

class1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2, scale=1, size=nClass1),
    'feature2': np.random.normal(loc=2, scale=1, size=nClass1),
    'target': [1] * nClass1 # number of one= 100
})
```

```python
class0, class1
```

```python
df=pd.concat([class0, class1]).reset_index(drop=True) 
df
# merge the two dataframes and then delete the duplicate one's
```

```python
df['target'].value_counts()
# this return the count of different values like zero and one here
```

# now we will do upsampling

```python
# we need to make minority and majority of dataframe
dfMinority=df[df['target']==1]
dfMajority=df[df['target']==0] # this target is equated to the value that we are getting like 900 zeros and 100 one
```

```python
dfMinority.head()
```

```python
dfMajority.head()
```

# now PERFORM UPSAMPLING-> for that we will need rescale library

```python
# now PERFORM UPSAMPLING-> for that we will need rescale libraryfrom sklearn.utils import resample
```

```python
# to upsample-> increase data point: FOR MINORITY
dfMinorityUpSample=resample(dfMinority, # which we have to upsample
                           replace=True, # implement sample with replacement, will add more additional sample
                            # downsampling me replace ko false krdena hai bas
                           n_samples=len(dfMajority), # how many sample i have to make: TO MATCH THE MAJORITY CLASS
                           random_state=42 # ANY value
                           )

# n_sample aur random_state, if we make a mistake in this name; we will face an error because this library is very strict on its naming
```
