```python
#Evidence to say that the two population variances are not equal.

worker1= [18, 19, 22, 25, 27, 28, 41, 45, 51, 55]
worker2= [14, 15, 15, 17, 18, 22, 25, 25, 27, 34]
# using f test we can find out that which worker is performing well
```

```python
## Calaculating Ftest
import numpy as np
f_test=np.var(worker1)/np.var(worker2) # variance ratio both the worker
f_test
```

```python
## Degree OF fReedom
df1=len(worker1)-1
df2=len(worker2)-1
signifance_value=0.05

import scipy.stats as stat

# finding crtical value -> we will need degree of freedom of both worker
critical_value = stat.f.ppf(q=1-signifance_value,dfn=df1,dfd=df2)

# q=1-signifance_value: THIS IS THE CONFIDENCE INTERVAL
# DFN: DEGREE OF FREEDOM OF NUMERATOR
```
