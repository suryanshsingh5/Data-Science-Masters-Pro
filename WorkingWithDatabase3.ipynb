```python
def insertDocument():
    studentInfo = {
        "name": "Drake",
        "section": 2,
        "maths_marks": 35,
        "sst_marks": 79
    }
    student_id = collection.insert_one(studentInfo).inserted_id
    print(f"Student with id {student_id} has been created")

insertDocument()
```

```python
def readDocuments():
    # Inserting a Document
    
    # Reading a Collection
    # Using find function 
    myStudents = collection.find({"section": 1, "name": "Harry"})
    
    # print(myStudents)
    for student in myStudents:
        print(student)
        
    # Using findOne function 
    myStudent = collection.find_one({"section": 1})
    print(myStudent) 

readDocuments()
```

```python
def updateDocuments():
    # using update one function-> increament of section 1 to 100; takes 2 argument
    collection.update_one({"section": 1}, {'$inc': {'section': 100}})

    # using update many function-> increament section of all
    collection.update_many({}, {'$inc': {'section': 100}})

updateDocuments()
```

```python
def deleteDocuments():

    # delete one funciton
    r = collection.delete_one({"section": 102})
    print(r) # the object which is deleted
    print(r.deleted_count) # number of object deleted

    #delete many
    r = collection.delete_many({})
    print(r)
    print(r.deleted_count)
    collection.update_many({})
    
```
