# Measure of Dispersion

```python
ages_lst=[23,43,23,56,74,32,68,98,45,32]
```

```python
import numpy as np
```

```python
# calculating mean
np.mean(ages_lst)
```

```python
# calculating variance
np.var(ages_lst)
```

```python
#standard deviation
np.std(ages_lst)
```

```python
# plotting histogram using seaborn
import seaborn as sns
sns.histplot(ages_lst, kde=True)
```
